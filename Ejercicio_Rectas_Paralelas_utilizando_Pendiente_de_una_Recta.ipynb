```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definición de los puntos
P1 = (0, 3)
P2 = (1, -1)
P3 = (1, 2)
P4 = (0, 6)

# Cálculo de las pendientes
m1 = (P2[1] - P1[1]) / (P2[0] - P1[0])
m2 = (P4[1] - P3[1]) / (P4[0] - P3[0])

# Crear un rango de valores de x para graficar la línea
x_values = np.linspace(-2, 2, 400)
y_values1 = m1 * (x_values - P1[0]) + P1[1]
y_values2 = m2 * (x_values - P3[0]) + P3[1]

# Configuración y graficación
plt.figure(figsize=(8,8))
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.scatter([P1[0], P2[0]], [P1[1], P2[1]], color='red')  # Puntos para la primera línea
plt.scatter([P3[0], P4[0]], [P3[1], P4[1]], color='blue')  # Puntos para la segunda línea
plt.plot(x_values, y_values1, label=f'Línea a través de P1 y P2: m1={m1}', color='red')
plt.plot(x_values, y_values2, label=f'Línea a través de P3 y P4: m2={m2}', color='blue')
plt.legend()
plt.show()
```
