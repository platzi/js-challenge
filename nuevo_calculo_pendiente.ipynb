```python
import matplotlib.pyplot as plt
import numpy as np

# Funci칩n para calcular la pendiente entre dos puntos
def calcular_pendiente(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)

# Coordenadas de dos puntos
x1, y1 = -3, 7
x2, y2 = 3, 0

# Calcular la pendiente
pendiente = calcular_pendiente(x1, y1, x2, y2)

# Puntos en el eje x para dibujar la recta
x = np.linspace(-10, 10, 400)
y = pendiente * (x - x1) + y1  # Ecuaci칩n de la recta en la forma punto-pendiente: y - y1 = m(x - x1)

# Dibujar el gr치fico
plt.figure(figsize=(8, 8))
plt.plot(x, y, '-r', label=f'Pendiente = {pendiente}')
plt.scatter([x1, x2], [y1, y2], color='blue')  # Dibuja los puntos dados
plt.title('Gr치fico de una Recta')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
